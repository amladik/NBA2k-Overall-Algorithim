**Importing the Training and Validation Data**


```python
nba_data = pd.read_csv('../input/nba-2k-ratings-with-real-nba-stats/nba_rankings_2014-2020.csv')
nba_data.drop(['Unnamed: 0'], axis=1, inplace=True)

nba_data.head()


```

**Importing the Test Data**

```python
test_data = pd.read_csv('../input/titololo/testData.csv')
test_data.drop(['Unnamed: 0'], axis=1, inplace=True)
test_data
```

```python
srs_data = pd.read_csv('../input/srsdata/srstestdatanba.csv') 
srs_data.drop(['Unnamed: 0'], axis=1, inplace=True)
srs_data
```

```python
#import statements
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
#define label to be rating
y = nba_data.rankings
#features are all stats

featureList = ['GP', 'AGE','W', 'L', 'MIN', 'PTS', 'FGM', 'FGA', 'FG%', '3PM', '3PA', '3P%',
       'FTM', 'FTA', 'FT%', 'OREB', 'DREB', 'REB', 'AST', 'TOV', 'STL', 'BLK',
       'PF', '+/-' ]


x = nba_data[featureList]
test_x = test_data[featureList]
srs_x = srs_data[featureList]
print(test_x)
```

```python
#create validation data

train_x, val_x, train_y, val_y = train_test_split(x, y, random_state=1)
```

```python
#create model
model= RandomForestRegressor(n_estimators = 1000, max_depth =7.5,random_state = 1)
model.fit(train_x,train_y)

```

```python
validationPredictions = model.predict(val_x)
print ('Model Guess:')
val = 65
print((int(validationPredictions[val]))*1.0)
print(val_x.iloc[[val]])
print(val_y.iloc[[val]])

#print(nba_data['PLAYER'].iloc[[val]])
mse = mean_squared_error(validationPredictions , val_y)
```
